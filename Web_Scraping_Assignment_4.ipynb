#### Q. No. 2:-
- Scrape the details team India’s international fixtures from bcci.tv.
- Url = https://www.bcci.tv/.
- You need to find following details:
    - Match title (I.e. 1st ODI)
    - Series
    - Place
    - Date
    - Time
#### Note: - 
From bcci.tv home page you have reach to the international fixture page through code.


```python
# Loading the web driver

driver=webdriver.Chrome("chromedriver.exe") 
```

```python
# Loading the web-page

url2 = 'https://www.bcci.tv/'

driver.get(url2)
```

```python
# Visiting the international fixture page.

fix_tag = driver.find_element_by_xpath('//div[@class="navigation__drop-down drop-down drop-down--reveal-on-hover"]')

fix_tag.click()

fix_tag2 = driver.find_element_by_xpath('//a[@class="navigation__link navigation__link--in-drop-down"]')
fix_tag2.click()
```

```python
# Creating empty list
match_urls = []
match_title = []
series = []
place_match = []
date_match = []
time_match = []
```

```python
# Scraping URLs

mtch_url = driver.find_elements_by_xpath('//a[@class="js-match event-list__item event-list-item fixture  t-test fixture--long event-list-item--long"]')

for u in mtch_url:
    match_urls.append(u.get_attribute('href'))
match_urls
```

```python
# Scraping the match title
mth = driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]')
for m in mth:
    try:
        match_title.append(m.text)
    except NoSuchElementException:
        match_title.append('--')
time.sleep(5)

# Scraping the series name
ser = driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]')
for s in ser:
    try:
        series.append(s.text)
    except NoSuchElementException:
        series.append('--')

time.sleep(5)

# Scraping the match place

plc = driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]//span')
for p in plc:
    try:
        place_match.append(p.text)
    except NoSuchElementException:
        place_match.append('--')
        
time.sleep(5)

# Scraping match time
tim_tag = driver.find_elements_by_xpath('//span[@class="fixture__time"]')
for t in tim_tag:
    try:
        time_match.append(t.text)
    except NoSuchElementException:
        time_match.append('--')
time.sleep(5)

# Scraping the match date
for u2 in match_urls:
    driver.get(u2)
    
    dte = driver.find_elements_by_xpath('//div[@class="mc-header-scorebox__datetime"]//strong')
    for d in dte:
        try:
            date_match.append(d.text)
        except NoSuchElementException:
            date_match.append('--')
```

```python
# Checking the length of the data which we have scraped.

print (len(match_title),len(series),len(place_match),len(time_match),len(date_match))
```

```python
# Making dataframe:

fixtures = pd.DataFrame({})

fixtures['Match Title'] = match_title
fixtures['Series'] = series
fixtures['Place'] = place_match
fixtures['Date'] = date_match
fixtures['Time'] = time_match

fixtures
```

```python
# Creating empty list

url_respository = []
respository_title = []
description_respository =[]
count_contributor = []
used_language = []
```

```python
# Scraping URLs
url_tag5 = driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]//a')
for u in url_tag5:
    url_respository.append(u.get_attribute('href'))
url_respository
```

```python
len(url_respository)
```

```python
# Scraping respository title
title_tag5 = driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]')
for t in title_tag5:
    try:
        respository_title.append(t.text.replace('/',''))
    except NoSuchElementException:
        respository_title.append('--')
respository_title
```

```python
len(respository_title)
```

```python
# Scraping contributer count
count_tag5 = driver.find_elements_by_xpath('//a[@class="Link--muted d-inline-block mr-3"]')
data5 = []
for c in count_tag5:
    try:
        data5.append(c.text)
    except NoSuchElementException:
        data5.append('--')
time.sleep(5)        

for i5 in range (0, len(data5)):
    if i5%2:
        count_contributor.append(data5[i5])
    
count_contributor
```

```python
len(count_contributor)
```

```python
# Scraping respository description
for u in url_respository:
    driver.get(u)
    
    desc_tag5 = driver.find_elements_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div')
    for d in desc_tag5:
        if d.text is None:
            description_respository.append('--')
        else:
            description_respository.append(d.text.replace('\n',''))
        
description_respository
```

```python
len(description_respository)
```

```python
# Scraping language used

for u in url_respository:
    driver.get(u)
    lan_tag5 = driver.find_elements_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div/div/ul/li[1]/a/span[1]')
    if lan_tag5:
        for l in lan_tag5:
            used_language.append(l.text.replace('\n',''))
    else:
        used_language.append('No languages used')
        

used_language
```

```python
len(used_language)
```

```python
github = pd.DataFrame({'Respository Title':respository_title, 'Respository Description':description_respository,
                       'Contributor Count':count_contributor, 'Language Used':used_language})

github
```
